```python
# Importar librerias
import spacy
import pandas as pd
```

```python
#string
text="El Real Madrid es el mejor equipo de la historia y en el mismo esta jugando Vinicius Junior quien es un gran jugador brasileño"
```

```python
#Descarga del pipeline
!python -m spacy download es_core_news_sm -q
```

```python
#cargar pipelines
nlp=spacy.load("es_core_news_sm")
```

```python
#crear objeto coumento
doc=nlp(text)
```

```python
for token in doc:
    print(token.text,"     ",token.is_stop,"     ",token.is_punct)
```

```python
def token_stop(documento):
    lista_token=[]
    lista_stop=[]
    lista_lema=[]
    lista_pos=[]

    for token in documento:
        if not token .is_stop and not token.is_punct and not token.is_digit and not token.like_url and not token.like_email:
            lista_token.append(token.text)
            lista_stop.append(token.is_stop)
            lista_lema.append(token.lemma_)
            lista_pos.append(token.pos_)
    
    df=pd.DataFrame(list(zip(lista_token,lista_stop,lista_lema,lista_pos)), columns=["Toke","Stopword","Lemma","Poss"])
    return df
```

```python
token_stop(doc)
```

```python
#vizualizacion
from spacy import displacy
displacy.render(doc,style="dep", jupyter=True)
```
