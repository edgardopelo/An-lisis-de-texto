```python
def token_stop(documento):
   lista_token=list()
   lista_stop=list()
   lista_lema=list()
   lista_pos=list()
   for token in documento:
     if not token.is_stop and not token.is_punct and not token.like_url and not token.is_digit and not token.is_quote:
       lista_token.append(token.text)
       lista_stop.append(token.is_stop)
       lista_lema.append(token.lemma_)
       lista_pos.append(token.pos_)
    
   df = pd.DataFrame(list(zip(lista_token,lista_stop,lista_lema, lista_pos)), columns = ['Token','Stopword',"Lemma","Poss"])
   return df
```

```python
def remove_emails_url(documento):
  #Otros atributos de los tokens (Reconocimiento de Emails)
  lista_tokensd=[]

  for token in documento:
    if token.like_email or  token.like_num or token.like_url:
      continue
  
    else:
      lista_tokensd.append(token)
  #print(lista_tokensd) 
  return lista_tokensd

```

```python
import re
def remove_emoji(string):#####Funci√≥n para eliminar simbolor raros. Lo otro se deja a spacy.
    emoji_pattern = re.compile("["
                u"\U0001F600-\U0001F64F"  
                u"\U0001F300-\U0001F5FF"  
                u"\U0001F680-\U0001F6FF"  
                u"\U0001F1E0-\U0001F1FF"  
                u"\U00002702-\U000027B0"
                u"\U000024C2-\U0001F251"
                u"\U0001f926-\U0001f937"
                u'\U00010000-\U0010ffff'
                u"\u200d"
                u"\u2640-\u2642"
                u"\u2600-\u2B55"
                u"\u23cf"
                u"\u23e9"
                u"\u231a"
                u"\u3030"
                u"\ufe0f"
    "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', string)
```

```python
limpiando_emojis=[]
for i in df["Tweet"]:
    limpiando_emojis.append(remove_emoji(i))
df["Tweet"]=limpiando_emojis
df
```
